```python
import pyart
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np

filename1 = 'PARstrat/output/20090527/cfrad.20090527_145347.000_to_20090527_145735.000_PAR_v1_SUR.nc'
filename2 = 'PARconv/output/20091118/cfrad.20091118_062002.000_to_20091118_062342.998_PAR_v1_SUR.nc'
```

```python
RadParStrat = pyart.io.read(filename1)
RadParConv  = pyart.io.read(filename2)
```

```python
print RadParConv.fields.keys()
```

```python
display = pyart.graph.RadarMapDisplay(RadParConv)

f = plt.figure(figsize=[15,10])

plt.subplot(2, 2, 1) 
display.plot_ppi_map('dBZ', sweep=0, vmin = -10, vmax = 60,
                     lat_lines = np.arange(-35,-30,1.), lon_lines = np.arange(-65,-55,1.))
display.plot_range_ring(240., line_style='k-')

plt.subplot(2, 2, 2) 
display.plot_ppi_map('ZDR', sweep=0, vmin = -3, vmax = 6,
                    lat_lines = np.arange(-35,-30,1.), lon_lines = np.arange(-65,-55,1.))
display.plot_range_ring(240, line_style='k-')

plt.subplot(2, 2, 3) 
display.plot_ppi_map('RhoHV', sweep=0, vmin = 0.9, vmax = 1,
                    lat_lines = np.arange(-35,-30,1.), lon_lines = np.arange(-65,-55,1.))
display.plot_range_ring(240., line_style='k-')

plt.subplot(2, 2, 4) 
display.plot_ppi_map('uPhiDP', sweep=0, vmin = 0, vmax = 360,
                    lat_lines = np.arange(-35,-30,1.), lon_lines = np.arange(-65,-55,1.))
display.plot_range_ring(240., line_style='k-')
```

```python
f = plt.figure(figsize=[16,12])

display = pyart.graph.RadarDisplay(RadParStrat)

# Reflectividad horizontal caso estratiforme
plt.subplot(2, 2, 1) 
display.plot_ppi('dBZ', sweep=0, vmin = -10, vmax = 60)
display.plot_range_rings([120])

# PhiDP caso estratiforme
plt.subplot(2, 2, 2) 
display.plot_ppi('uPhiDP', sweep=0, vmin = 0, vmax = 360)
display.plot_range_rings([120])

display = pyart.graph.RadarDisplay(RadParConv)

# Reflectividad horizontal caso convectivo
plt.subplot(2, 2, 3) 
display.plot_ppi('dBZ', sweep=0, vmin = -10, vmax = 60)
display.plot_range_rings([120])

# PhiDP caso convectivo
plt.subplot(2, 2, 4)
display.plot_ppi('uPhiDP', sweep=0, vmin = 0, vmax = 360)
display.plot_range_rings([120])
```
